Total de unidades vendidas por país de origen del vendedor. Calcular el total. ¿Qué países tienen una mayor y menor participación en las ventas totales? 

```python
import pandas as pd
import numpy
```

```python
df = pd.read_csv('merchant_fill.csv') # cargamos el DF de trabajo
pd.set_option('display.max_columns', None) # esto sirve para cargar todas las columnas en la visualización del DF
```

```python
df.head() # comprobamos que se ha cargado bien
```

```python
df.columns # esto lo usamos para ver los nombres de las columnas y poder hacer las consultas
```

```python
df_total_unidades_pais = df.groupby(["origin_country", "units_sold"]).count().reset_index() # el problema aquí es que se agrupan las dos cosas juntas, de manera que cada vez que
                                                                                            # units_sold cambia, lo trata como una cosa que contar
df_total_unidades_pais
```

```python
df_total = df.groupby("origin_country")['units_sold'].count().reset_index() # esto cuenta cuántas veces aparece el país
df_total
```

```python
df_total_sum = df.groupby("origin_country")['units_sold'].sum().reset_index() # esto suma todas las unidades vendidas
df_total_sum
```

```python
df_total_sum.sort_values(by='units_sold', ascending=False) # usamos sort_values para ordenarlo por lo que determinamos en el by; después, ascending=False porque ascending es un booleano
```

```python
df.columns # esto lo usamos para ver los nombres de las columnas y poder hacer las consultas
```

```python
estadisticas_precio_mean = df.groupby('countries_shipped_to')['price'].mean()
estadisticas_precio_mean.tail()
```

```python
estadisticas_precio_max = df.groupby('countries_shipped_to').agg(
    promedio = ('price', 'mean'),
    maximo = ('price', 'max')
).reset_index()

estadisticas_precio_max.tail()
```
